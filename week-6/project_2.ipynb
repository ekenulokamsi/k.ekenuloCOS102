```python
class Delivery:
    def __init__(self, location, weight):
        self.location = location.lower() # Convert to lowercase for easier matching
        self.weight = weight
        self.amount = 0

    def calculate_fee(self):
        # Condition for PAU
        if self.location == "pau":
            if self.weight >= 10:
                self.amount = 2000
            else:
                self.amount = 1500
        
        # Condition for Epe
        elif self.location == "epe":
            if self.weight >= 10:
                self.amount = 5000
            else:
                self.amount = 4000
        
        else:
            self.amount = 0 # Location not recognized

    def display_receipt(self):
        if self.amount > 0:
            print("-" * 30)
            print(f"Location: {self.location.upper()}")
            print(f"Weight: {self.weight}kg")
            print(f"Total Charge: N{self.amount}")
            print("-" * 30)
        else:
            print("Error: Location not recognized. Please enter 'PAU' or 'Epe'.")

# --- Execution ---
loc = input("Enter delivery location (PAU or Epe): ")
wgt = float(input("Enter package weight (kg): "))

package = Delivery(loc, wgt)
package.calculate_fee()
package.display_receipt()
```
