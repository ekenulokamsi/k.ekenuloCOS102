```python
import random

class Employee:
    # List of authorized employees and available tasks
    authorized_employees = [
        "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", "Andrew Ugwu", 
        "Stella Mankinde", "Jane Akibo", "Ago James", "Michell Taiwo", 
        "Abraham Jones", "Nicole Anide", "Kosi Korso", "Adele Martins", 
        "Emmanuel Ojo", "Ajayi Fatima"
    ]
    
    tasks = ["Loading", "Transporting", "Reviewing Orders", "Customer Service", "Delivering Items"]

    def __init__(self, name):
        self.name = name
        self.is_present = False
        self.assigned_task = None

    def check_employee(self):
        """Checks if the user is in the authorized list."""
        if self.name in self.authorized_employees:
            print(f"Access Granted. Welcome, {self.name}!")
            return True
        else:
            self.refuse_access()
            return False

    def take_attendance(self):
        """Marks the employee as present."""
        self.is_present = True
        print(f"Attendance recorded for {self.name}.")

    def assign_task(self):
        """Randomly assigns a task from the list."""
        # random.randint(0, len(list)-1) picks a random index
        random_index = random.randint(0, len(self.tasks) - 1)
        self.assigned_task = self.tasks[random_index]
        print(f"Task Assigned: {self.assigned_task}")

    def refuse_access(self):
        """Handles unauthorized users."""
        print("ACCESS DENIED: User not found in the employee database.")

# --- Running the Program ---

user_input = input("Please enter your name: ")
worker = Employee(user_input)

# Logic Flow
if worker.check_employee():
    worker.take_attendance()
    worker.assign_task()
```
