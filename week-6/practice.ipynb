```python
class Addition:
    # Constructor
    def __init__(self, f, s):
        self.first = f
        self.second = s
        self.answer = 0

    # Method to calculate
    def calculate(self):
        self.answer = self.first + self.second

    # Method to show results
    def display(self):
        print("First number = " + str(self.first))
        print("Second number = " + str(self.second))
        print("Addition of two numbers = " + str(self.answer))

# --- Executing the code ---
try:
    n1 = int(input("Enter first number: "))
    n2 = int(input("Enter second number: "))

    obj = Addition(n1, n2)
    obj.calculate()
    obj.display()
except ValueError:
    print("Please enter valid numbers!")
```

```python
class Dogs:
    # class attribute
    attr1 = "mammal"
    
    # Instance attribute
    def __init__(self, name):
        self.name = name

# Object instantiation
# Fix: Using 'Dogs' to match the class name above
dog1 = Dogs("Oscar")
dog2 = Dogs("Peaches")

# Accessing class attributes
# Fix: Using 'dog1' instead of 'dog3'
print("Oscar is a {}.".format(dog1.__class__.attr1))
print("Peaches is also a {}.".format(dog2.__class__.attr1))

# Accessing instance attributes
print("My name is {}.".format(dog1.name))
print("My name is {}.".format(dog2.name))
```

```python
class SST:
    # class attribute
    prog1 = "Computer Science"
    
    # Instance attribute
    def __init__(self, name):
        self.name = name
        
    def speak(self):
        print("My name is {}.".format(self.name))
        # Note: Use self.__class__.prog1 to access class attributes inside methods
        print("I'm studying {}.".format(self.__class__.prog1))

# Object instantiation
# Fix: Ensure parentheses and quotes are balanced
stud1 = SST(input("Enter the name of the first student: "))
stud2 = SST(input("Enter the name of the second student: "))

# Accessing class methods
stud1.speak()
stud2.speak()
```

```python
# Class for Dog
class Dog:
    # Class Variable
    animal = 'dog'

    # The init method or constructor
    def __init__(self, breed, color):
        # Instance Variable
        self.breed = breed
        self.color = color

# Objects of Dog class
# Fix: Use 'Dog' (singular) to match the class name above
Rodger = Dog("Pug", "brown")
Buzo = Dog("Bulldog", "black")

print('Rodger details:')
# Fix: Corrected 'aminal' to 'animal'
print('Rodger is a', Rodger.animal)
print('Breed: ', Rodger.breed)
print('Color: ', Rodger.color)

print('\nBuzo details:')
# Fix: Corrected 'aminal' to 'animal'
print('Buzo is a', Buzo.animal)
print('Breed: ', Buzo.breed)
print('Color: ', Buzo.color)

# Class variables can be accessed using class name also
print("\nAccessing class variable using class name")
print(Dog.animal)
```

```python
# Class for Dog
class Dog:
    # Class Variable
    animal = 'dog'

    # The init method or constructor
    def __init__(self, breed):
        # Instance Variable
        self.breed = breed
        self.color = None # Initialize color as empty until set

    # Adds an instance variable (Setter method)
    def setColor(self, color):
        self.color = color

    # Retrieves instance variable (Getter method)
    def getColor(self):
        return self.color

# Object instantiation
Rodger = Dog("pug")
Rodger.setColor("brown")

# Output the result
print(Rodger.getColor())
```
