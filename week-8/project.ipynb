```python
import tkinter as tk
from tkinter import messagebox

# Parent Class
class Zenith:
    def mutual_services(self):
        return ["Lines of credit", "Investment management and accounts", "Insurance"]

    def unique_services(self):
        pass   # Will be overridden (polymorphism)


# Child Classes
class Retail(Zenith):
    def unique_services(self):
        return [
            "Retirement and education accounts",
            "Loans and mortgages",
            "Checking and savings"
        ]


class Global(Zenith):
    def unique_services(self):
        return [
            "Multi-currency management",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"
        ]


class Commercial(Zenith):
    def unique_services(self):
        return [
            "Advisory services"
        ]


# Function to get services
def show_services():
    name = entry_name.get()
    division = entry_division.get().lower()

    if division == "retail":
        employee = Retail()
    elif division == "global":
        employee = Global()
    elif division == "commercial":
        employee = Commercial()
    else:
        messagebox.showerror("Error", "Invalid division")
        return

    services = employee.unique_services() + employee.mutual_services()

    result = f"Employee: {name}\nDivision: {division.capitalize()}\n\nServices:\n"
    result += "\n".join(services)

    output_label.config(text=result)


# GUI Setup
root = tk.Tk()
root.title("Zenith Bank Services")

tk.Label(root, text="Enter Name:").pack()
entry_name = tk.Entry(root)
entry_name.pack()

tk.Label(root, text="Enter Division (Retail/Global/Commercial):").pack()
entry_division = tk.Entry(root)
entry_division.pack()

tk.Button(root, text="Show Services", command=show_services).pack()

output_label = tk.Label(root, text="", justify="left")
output_label.pack()

root.mainloop()
```
