```python
import tkinter as tk
from tkinter import messagebox
import csv

# Function to check employee
def check_employee():
    name = name_entry.get().strip().lower()
    dept = dept_entry.get().strip().lower()

    if name == "" or dept == "":
        messagebox.showwarning("Input Error", "Please enter both name and department.")
        return

    found = False
    members = []

    try:
        with open("GIG-logistics.csv", "r") as file:
            reader = csv.DictReader(file)

            for row in reader:
                csv_name = row["Name"].strip().lower()
                csv_dept = row["Department"].strip().lower()

                # Collect members of same department
                if csv_dept == dept:
                    members.append(row["Name"])

                # Check if employee exists
                if name == csv_name and dept == csv_dept:
                    found = True

        if found:
            result_text.set(
                f"✅ Welcome {name.title()}!\n\n"
                f"Department Members:\n" + "\n".join(members)
            )
        else:
            result_text.set("❌ Employee does not exist.")

    except FileNotFoundError:
        messagebox.showerror("Error", "GIG-logistics.csv file not found!")

# ---------------- GUI ---------------- #

root = tk.Tk()
root.title("Employee Verification System")
root.geometry("400x400")

# Title
tk.Label(root, text="Employee Verification", font=("Arial", 14, "bold")).pack(pady=10)

# Name input
tk.Label(root, text="Enter Name:").pack()
name_entry = tk.Entry(root, width=30)
name_entry.pack(pady=5)

# Department input
tk.Label(root, text="Enter Department:").pack()
dept_entry = tk.Entry(root, width=30)
dept_entry.pack(pady=5)

# Button
tk.Button(root, text="Check Employee", command=check_employee).pack(pady=10)

# Result display
result_text = tk.StringVar()
tk.Label(root, textvariable=result_text, wraplength=350, justify="left").pack(pady=10)

# Run app
root.mainloop()
```
